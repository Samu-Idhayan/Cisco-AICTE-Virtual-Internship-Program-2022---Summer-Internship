```python
# re module provides support for regular expressions
import re

# Make a regular expression for validating an Ip-address
regex = "^((25[0-5]|2[0-4][0-9]|1[0-9][0-9]|[1-9]?[0-9])\.){3}(25[0-5]|2[0-4][0-9]|1[0-9][0-9]|[1-9]?[0-9])$"

all_ipv4_list = []
def getIp():
  present_ipv4 = input()
  if check(present_ipv4) != False:
    all_ipv4_list.append(present_ipv4)
  else:
    print("\nThis is an Invalid IP address...")
    print("Please, enter valid IP address: \n")
    return getIp()

def check(ip):
  # pass the regular expression
  # and the string in search() method
  if(re.search(regex, ip)):
    return True    
  else:
    return False

def convertIpAddr():
  file = open("conversion.txt", "w")
  for item in all_ipv4_list:
    temp_dec = []
    temp_bin = []
    temp_oct = []
    temp_hex = []
    parts = item.split(".")
    
    for part in parts:
      temp_dec.append(str(part))
      temp_bin.append(str(bin(int(part)).replace("0b","")))
      temp_oct.append(str(oct(int(part)).replace("0o","")))
      temp_hex.append(str(hex(int(part)).replace("0x","")))
    writeInFile(file, item, temp_dec, temp_bin, temp_oct, temp_hex)
    printResult(item, temp_dec, temp_bin, temp_oct, temp_hex)

def writeInFile(file, item, dec_list, bin_list, oct_list, hex_list):
  t1 = "IP Address: " + item
  file.write(t1+"\n")
  
  t2= "Decimal format: " + ".".join(dec_list)
  file.write(t2+"\n")
  
  t3 = "Binary format: " + ".".join(bin_list)
  file.write(t3+"\n")
  
  t4 = "Octal format: " + ".".join(oct_list)
  file.write(t4+"\n")
  
  t5 = "Hexadecimal format: " + ".".join(hex_list)
  file.write(t5+"\n"+"\n")

def printResult(item, dec_list, bin_list, oct_list, hex_list):
  t1 = "\nIP Address: " + item
  print(t1)

  t2= "Decimal format: " + ".".join(dec_list)
  print(t2)

  t3 = "Binary format: " + ".".join(bin_list)
  print(t3)

  t4 = "Octal format: " + ".".join(oct_list)
  print(t4)

  t5 = "Hexadecimal format: " + ".".join(hex_list)
  print(t5)

if __name__ == '__main__' :

  print("Enter 10 IPv4 address:")
  for i in range(10):
    getIp()

  convertIpAddr()
```
